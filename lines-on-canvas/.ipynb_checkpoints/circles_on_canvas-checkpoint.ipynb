```python
import numpy as np
import random
import shapefile
import pandas as pd
import matplotlib.pyplot as plt
import json
%matplotlib notebook

# change font
plt.rcParams['font.sans-serif'] = "Lato"
plt.rcParams['font.weight'] = "regular"
plt.rcParams['font.size'] = 12
plt.rcParams['axes.labelweight'] = 'bold'
plt.rc('text', usetex=True)
```

```python
def draw_circle(xi,yi,ri,c):
    
    # select a value for node thickness
    # random
    #thick = random.random()*5
    # beta distribution
    thick = 10*np.random.beta(1,5)
    # according to the radius
    #thick = min(10,1/ri)
    
    # draw top half of circle
    x = np.linspace(xi-ri,xi+ri,300)
    y = lambda x: np.sqrt(ri**2-(x-xi)**2) + yi
    plt.plot(x,y(x),color=c,lw=thick)

    # draw bottom half of circle
    x = np.linspace(xi-ri,xi+ri,300)
    y = lambda x: -np.sqrt(ri**2-(x-xi)**2) + yi
    plt.plot(x,y(x),color=c,lw=thick)
```

# number of circles to create

```python
points = 30
```

# try different color schemes

```python
plt.figure(figsize=(10,10))

plt.fill_between([0,100],0,100,fc='#f4b6c2')
for i in range(points):
    
    xi = 100*round(random.random(),1)
    yi = 100*round(random.random(),1)
    ri = random.randint(5,40)
    draw_circle(xi,yi,ri,'#eee3e7')

plt.xlim(10,90)
plt.ylim(10,90)
plt.xticks([])
plt.yticks([])
plt.tight_layout()
plt.savefig('figs/pink_1.png',dpi=300,bbox_inches=None, pad_inches=0.0)
plt.show()
```
